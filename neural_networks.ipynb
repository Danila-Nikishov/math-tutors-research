```python
# ===== Install dependencies =====
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense, Input, Dropout
from keras.optimizers import Adam
from sklearn.model_selection import train_test_split
```

```python
# ===== Definition of network architecture =====
epochs = 1000
batch_size = 32
learning_rate = 0.001

def create_model(input_dim):
    model = Sequential([
        Input(shape=(input_dim,)),
        Dense(64, activation='relu'),
        Dropout(0.1),
        Dense(32, activation='relu'),
        Dropout(0.1),
        Dense(16, activation='relu'),
        Dense(5, activation='softmax')
    ])
    model.compile(
        optimizer=Adam(learning_rate=learning_rate),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    return model
```

```python
# ===== Training with different embeddings =====
embeddings_info = [
    {"tokenization": "BPE_2500", 'embedding_type': 'Word2Vec_with_training',
     "path": "embeddings\\BPE_2500_with_training_embeddings.npy"},
    {"tokenization": "BPE_2500", 'embedding_type': 'Navec_pretrained',
      "path": "embeddings\\BPE_2500_pretrained_embeddings.npy"},
    {"tokenization": "BPE_4000", 'embedding_type': 'Word2Vec_with_training',
      "path": "embeddings\\BPE_4000_with_training_embeddings.npy"},
    {"tokenization": "BPE_4000", 'embedding_type': 'Navec_pretrained',
      "path": "embeddings\\BPE_4000_pretrained_embeddings.npy"},
    {"tokenization": "Unigram_2500", 'embedding_type': 'Word2Vec_with_training',
      "path": "embeddings\\Unigram_2500_with_training_embeddings.npy"},
    {"tokenization": "Unigram_2500", 'embedding_type': 'Navec_pretrained',
      "path": "embeddings\\Unigram_2500_pretrained_embeddings.npy"},
    {"tokenization": "Unigram_4000", 'embedding_type': 'Word2Vec_with_training',
      "path": "embeddings\\Unigram_4000_with_training_embeddings.npy"},
    {"tokenization": "Unigram_4000", 'embedding_type': 'Navec_pretrained',
      "path": "embeddings\\Unigram_4000_pretrained_embeddings.npy"}
]

results_df = pd.DataFrame(columns=[
    "tokenization", "embeddings", "val_accuracy", "val_loss"
])
y = np.load("data\\target.npy")

for emb in embeddings_info:
    print(f"\n==== Training on embedding: {emb['tokenization'] + ' ' + emb['embedding_type']} ====")
    
    X = np.load(emb["path"])
    X_train, X_val, y_train, y_val = train_test_split(
        X, y, test_size=0.1, random_state=42, stratify=y
    )
    model = create_model(X.shape[1])
  
    history = model.fit(
        X_train, y_train,
        validation_data=(X_val, y_val),
        epochs=epochs,
        batch_size=batch_size,
        verbose=1
    )
    
    results_df = pd.concat([results_df, pd.DataFrame([{
        "tokenization": emb["tokenization"],
        "embeddings": emb["embedding_type"],
        "val_accuracy": history.history['val_accuracy'][-1],
        "val_loss": history.history['val_loss'][-1],
    }])], ignore_index=True)
  
results_df.to_csv("embedding_training_results.csv", index=False)
```
