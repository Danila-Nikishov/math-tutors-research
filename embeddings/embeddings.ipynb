```python
# ===== Install dependencies =====
from gensim.models import Word2Vec
import numpy as np
from navec import Navec
```

```python
path = 'C:\\Python_Projects\\FCS_2025-2027\\NOD_2026\\Spring\\NLP_on_Python\\Mini_Project\\navec_hudlit_v1_12B_500K_300d_100q.tar'
navec = Navec.load(path)
```

```python
def sentence_embedding_weighted(sentence, model):
    vector = np.zeros(model.vector_size)
    weight_sum = 0.0
    for word in sentence:
        if word in model.wv:
            count = model.wv.get_vecattr(word, "count")
            weight = 1 / (count + 1)
            vector += weight * model.wv[word]
            weight_sum += weight
    if weight_sum == 0:
        return np.zeros(model.vector_size)
    return vector / weight_sum


def sentence_embedding_weighted_navec(sentence, model):
    vector_dim = navec['привет'].shape[0]
    vector = np.zeros(vector_dim, dtype=np.float32)
    count = 0
    for word in sentence:
        if word in model:
            vector += model[word]
            count += 1
    if count > 0:
        vector /= count
    return vector
```

## 1) SentencePiece BPE Tokens Processing

1.1) BPE 2500 tokens with training

```python
# ===== Download corresponding corpus =====
tokenized_corpus_11 = []
with open('C:\\Python_Projects\\FCS_2025-2027\\NOD_2026\\Spring\\NLP_on_Python\\Mini_Project'
          '\\tokenization\\tokenized_corpuses\\BPE_2500_tokenized_corpus.txt', 'r', encoding='utf-8') as f:
    for line in f:
        tokenized_corpus_11.append(line.strip().split())
```

```python
# ===== Train skip-gram Word2Vec model =====
model_11 = Word2Vec(
    vector_size = 32,            
    window=4,                  
    min_count=1,              
    sg=1,                      
    epochs=2000,
    seed=42,
    workers=4,
    negative=6,
)

model_11.build_vocab(tokenized_corpus_11) 

model_11.train(
    corpus_iterable=tokenized_corpus_11,
    total_examples=len(tokenized_corpus_11),
    epochs=50
)
```

```python
# ===== Compute sentence embeddings =====
BPE_2500_with_training_embeddings = np.array([
    sentence_embedding_weighted(sent, model_11)
    for sent in tokenized_corpus_11
])
BPE_2500_with_training_embeddings = BPE_2500_with_training_embeddings.astype(np.float32)
np.save("BPE_2500_with_training_embeddings.npy", BPE_2500_with_training_embeddings)
```
