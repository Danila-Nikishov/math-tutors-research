```python
# ===== Install dependencies ======
import sentencepiece as spm
import pandas as pd
import re
```

```python
# ===== Data preparation for SentencePiece =====
data = pd.read_csv('C:\\Python_Projects\\FCS_2025-2027\\NOD_2026\\Spring\\NLP_on_Python\\Mini_Project\\data\\processed_data.csv', index_col = False)

def normalize_whitespace(text):
    return re.sub(r'\s+', ' ', text).strip()

with open('corpus.txt', 'w', encoding='utf-8') as f:
    for text in data.Unified_description:
        clean_text = normalize_whitespace(text)
        f.write(clean_text + '\n')
```

## 1) BPE Tokenization with SentencePiece 

1.1. Vocabulary size = 2500

```python
spm.SentencePieceTrainer.train(
    input='C:\\Python_Projects\\FCS_2025-2027\\NOD_2026\\Spring\\NLP_on_Python\\Mini_Project\\data\\corpus.txt', 
    model_prefix='BPE_2500', vocab_size=2500, model_type = 'bpe',
    character_coverage = 0.9995, split_by_unicode_script = True)

sp = spm.SentencePieceProcessor(model_file='BPE_2500.model')

with open('BPE_2500_tokenized_corpus.txt', 'w', encoding='utf-8') as f:
    for text in data.Unified_description:
        clean_text = normalize_whitespace(text)
        tokenized_text = sp.encode(clean_text, out_type=str)
        f.write(' '.join(tokenized_text) + '\n')
```
